```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
fontdict = {'family': 'Times New Roman', 'size': 20}
import matplotlib.transforms as mtransforms
import matplotlib.colors as mcolors
```

```python
x = [0.0001, 0.0005, 0.001, 0.005, 0.025, 0.05, 0.075, 0.1]
y = [1e-7, 1e-6, 1e-5, 5e-5, 1e-4, 5e-4, 1e-3, 1e-2]
data = pd.read_csv("ULEAP_hyperparameter.csv", header=None)
z = data.iloc[:, 2].to_numpy()
z = z.reshape(len(x), len(y)).transpose()
fig, ax = plt.subplots(figsize=(10,10))
im = ax.imshow(np.log(z/3000))
ax.set_xticks(np.arange(len(x)), labels=x)
ax.set_yticks(np.arange(len(y)), labels=y)
for i in range(len(y)):
    for j in range(len(x)):
        value = z[i,j]/3
        text = ax.text(j, i, value.astype(int),
                       ha="center", va="center", color="w")
        
plt.tight_layout()
plt.savefig("MAPK_parameters.pdf", dpi=600)
```

```python
x = [0.0001, 0.0005, 0.001, 0.005, 0.025, 0.05, 0.075, 0.1] # bound_factors
y = [1e-7, 1e-6, 1e-5, 5e-5, 1e-4, 5e-4, 1e-3, 1e-2] # record_time_interval
lx = ["0.000100", "0.000500", "0.001000", "0.005000", "0.025000", "0.050000", "0.075000", "0.100000"] # bound_factors
ly = ["0.000000", "0.000001", "0.000010", "0.000050", "0.000100", "0.000500", "0.001000", "0.010000"] # record_time_interval
```

```python
count = 0
ne = []
na = []
ua0 = []
for i in lx:
    a = []
    e = []
    a0 = []
    for j in ly:       
        filename = "ULEAP_leap_size_" + i + "_" + j + ".csv"
        data = pd.read_csv(filename, header=0)
        xe = data.iloc[:,1].to_numpy() # total events
        xa = data.iloc[:,2].to_numpy() # accept events
        xa0 = data.iloc[:,3].to_numpy() # ua0
        e.append(xe.mean())
        a.append(xa.mean())
        a0.append(xa0.mean())
    na.append(a)
    ne.append(e)
    ua0.append(a0)
```

```python
colors = list(mcolors.TABLEAU_COLORS)
plt.figure(figsize=(10,8))

for i, v in enumerate(ne):
    plt.plot(np.log10(y), np.log10(v), '-', color=colors[i], label=r"$\bound_factors = $" + f"{x[i]}")
    
for i, v in enumerate(na):
    plt.plot(np.log10(y), np.log10(v), linestyle='dashed', color=colors[i])
    
for i, v in enumerate(ua0):
    plt.plot(np.log10(y), np.log10(np.array(y)*np.array(v)), linestyle='dashdot', color=colors[i])
    
plt.legend()
plt.xlabel(r"$\record_time_interval$", fontdict=fontdict)
plt.ylabel("Number of accepted events", fontdict=fontdict)
plt.tight_layout()
#plt.hlines(y=np.log10(500), xmin=-7, xmax=-2)
#plt.vlines(x=np.log10(0.0001), ymin=0, ymax=4)
plt.savefig("leap_size_and_efficiency.pdf", dpi=600)
```

```python
plt.figure(figsize=(10,8))

data = pd.read_csv("MAPK_benchmark.csv", header=None)
plt.plot(data.iloc[:,0], data.iloc[:, 1]/1000, 'ro-', lw=2, label='DM')
plt.plot(data.iloc[:,0], data.iloc[:, 2]/1000, 'c^-', lw=2, label='NRM')
plt.plot(data.iloc[:,0], data.iloc[:, 3]/1000, 'm>-', lw=2, label='RSSA')
plt.plot(data.iloc[:,0], data.iloc[:, 4]/1000, 'y<-', lw=2, label='USSA')
plt.plot(data.iloc[:,0], data.iloc[:, 5]/1000, 'gv-', lw=2, label='RNRM')
plt.plot(data.iloc[:,0], data.iloc[:, 6]/1000, 'b+-', lw=2, label='RLEAP')

plt.xlabel("Time interval", fontdict=fontdict)
plt.ylabel("CPU time (s)", fontdict=fontdict)
plt.tight_layout()

plt.legend()

plt.savefig("MAPK_benchmark.pdf", dpi=600)
```
