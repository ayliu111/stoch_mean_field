```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
fontdict = {'family': 'Times New Roman', 'size': 20}
import matplotlib.transforms as mtransforms
from scipy.special import comb
```

```python
model = "isomerization"
method = "blwileap"
```

```python
data_file = "../data/" + model + "/" + method + "/population_dist.csv"
```

```python
data_99 = pd.read_csv(data_file, header=0)
```

```python
data_8 = pd.read_csv(data_file, header=0)
```

```python
data_5 = pd.read_csv(data_file, header=0)
```

```python
data_1 = pd.read_csv(data_file, header=0)
```

```python
data_01 = pd.read_csv(data_file, header=0)
```

```python
data_001 = pd.read_csv(data_file, header=0)
```

```python
data_0001 = pd.read_csv(data_file, header=0)
```

```python
data = [data_99, data_8, data_5, data_1, data_01, data_001, data_0001]
label = ["0.99", "0.8", "0.5", "0.1", "0.01", "0.001", "0.0001"]
for i, d in enumerate(data):
    weight = d["Weight"].to_numpy()
    print(weight.max(), weight.min())
    h = plt.hist(np.log10(weight), bins=100, histtype="step", label="bw=" + label[i])

plt.legend()
plt.savefig("weight_dist_and_bound_weight.pdf", dpi=300)
```
