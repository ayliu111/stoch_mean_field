```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
fontdict = {'family': 'Times New Roman', 'size': 20}
import matplotlib.transforms as mtransforms
from scipy.optimize import fsolve
from scipy.stats import poisson
```

Let $h: [-1, \infty) \to \mathbb(R)$ be the function defined by $h(u) = 2 \cdot \frac{(1+u)\ln(1+u)-u}{u^2}$.

Let $X \sim Poisson(\lambda)$, for some parameter $\lambda > 0$. Then for any $x > 0$, we have

$Pr\{X \ge \lambda + x\} \le e^{-\frac{x^2}{2\lambda} h(\frac{x}{\lambda})}$. 

Suppose we know the maximum allowed number of events is $n$, then we demand the probability of chosing number of events larger than $n$ to be less than $p$. i.e. $Pr\{X \ge n\} \le p$. 

We thus set $\lambda + x = n$, so $Pr\{X \ge n\} = Pr\{X \ge \lambda + x\} \le e^{-\frac{x^2}{2\lambda} h(\frac{x}{\lambda})}$. 

$h(\frac{x}{\lambda}) = 2\lambda \cdot \frac{n(\ln n - \ln \lambda) - x}{x^2}$. And $e^{-\frac{x^2}{2\lambda} h(\frac{x}{\lambda})} = e^{(n \ln n - n) -(n \ln \lambda - \lambda)}$.

Thus we see $Pr\{X \ge n\} \le e^{(n \ln n - n) -(n \ln \lambda - \lambda)}$. By setting $e^{(n \ln n - n) -(n \ln \lambda - \lambda)} = p$, we find 

$n\ln \lambda - \lambda = \ln p + n \ln n - n$.

```python
def func(x, n, p):
    return n * np.log(x) - x - np.log(p) - n * np.log(n) + n
```

```python
x = np.linspace(1, 2000, 1000)
p = 0.01
n = 1000
y = np.log(p) + n * np.log(n) - n
plt.plot(x, n * np.log(x) - x)
plt.plot(x, y * np.ones(1000))
```

```python
root = fsolve(func, 4, args=(n, p))
root
```

```python
mu = root
x = np.arange(0, poisson.ppf(0.9999, mu))
plt.plot(x, poisson.pmf(x, mu), 'bo', ms=8, label='poisson pmf')
plt.plot(x, poisson.cdf(x, mu), 'ro', ms=8, label='poisson cdf')
plt.plot(x, 1 - poisson.cdf(x, mu), 'go', ms=8, label='1 - cdf')
plt.legend()
```

```python
mu = root
poisson.ppf(1-p, mu)
```

We also have $Pr\{X \ge \lambda + x\} \le e^{- \frac{x^2}{2(\lambda + x)}}$, setting $e^{- \frac{x^2}{2(\lambda + x)}} = p$, we have a simpler form for determining $\lambda$:

$\lambda = n - \sqrt{- 2 n \ln p}$

```python
plt.figure(figsize=(10,8))
seq_n = np.arange(10, 1000)
x1 = []
x2 = []
diff = []
p = 0.0001
for i in seq_n:
    t1 = fsolve(func, 1, args=(i, p))
    x1.append(t1)
    t2 = i - np.sqrt(-2 * i * np.log(p))
    x2.append(t2)
    diff.append(t1 - t2)
plt.plot(seq_n, x1, 'r', label="accurate")
plt.plot(seq_n, x2, 'b', label="approx")
plt.plot(seq_n, diff, 'g', label="diff")
plt.ylabel(r"$\lambda$", fontsize=20)
plt.xlabel(r"$n$", fontsize=20)
plt.legend()
print(np.log(p), np.log(p)/2, np.sqrt(-np.log(p)))
print(diff[0], diff[-1])
```

```python
plt.figure(figsize=(10,8))
n = 10
p = np.linspace(1e-5, 0.9, 1000)
x1 = []
x2 = []
diff = []
x3 = []
x4 = []
for i in p:
    t1 = fsolve(func, 1, args=(n, i))
    x1.append(t1)
    t2 = n - np.sqrt(-2 * n * np.log(i))
    x2.append(t2)
    diff.append(t1 - t2)
    t3 = n - np.sqrt(-2 * n * np.log(i)) - np.log(i)/1.78
    x3.append(t3)
    t4 = - np.log(i) / (t1 - t2)
    x4.append(t4)
plt.plot(p, x1, 'r', label="accurate")
plt.plot(p, x2, 'b', label="approx")
plt.plot(p, diff, 'g', label="diff")
plt.plot(p, -np.log(p)/1.7, 'k', label=r"$\ln p / 1.78$")
plt.plot(p, x3, 'y', label="corrected")
plt.plot(p, x4, 'k', label="error")
plt.xlabel(r"$p$", fontsize=20)
plt.ylabel(r"$\lambda$", fontsize=20)
plt.legend()
plt.title(r"$n = 10$")
```

```python
plt.figure(figsize=(10,8))
n = [10, 100, 1000, 10000]
p = np.linspace(1e-5, 0.9, 1000)
for i in n:
    x1 = []
    x2 = []
    for j in p:
        t1 = fsolve(func, 1, args=(i, j))
        x1.append(t1)
        t2 = i - np.sqrt(-2 * i * np.log(j)) - np.log(j) / 1.78
        x2.append(t2)
    plt.plot(p, np.array(x1)/i, 'o', ms=2, label=rf"$n$ = {i}")
    plt.plot(p, np.array(x2)/i, 'k-')
    
plt.xlabel(r"$p$", fontsize=20)
plt.ylabel(r"$\lambda$", fontsize=20)
plt.legend()
plt.title(r"approximate of $\lambda$, results normalized by $n$")
```

### Conclusion

The difference between root of $n \ln \lambda - \lambda = \ln p + n \ln n - c$ and $n - \sqrt{-2 n \ln p}$ is decreasing along with $p$, and insensitive with $n$. 
Thus calculate the difference at a small $p$ is sufficient to modify the approximate result. 
Here we calculate the $\lambda = n - \sqrt{-2 n \ln p} - \frac{1}{1.78}\ln p$
