```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
fontdict = {'family': 'Times New Roman', 'size': 20}
```

```python
plt.figure(figsize=(10,8))
dm = pd.read_csv("DM_MAPK296_trajectory.csv", header=0) 
ugleap = pd.read_csv("UGLEAP_MAPK296_trajectory.csv", header=0) 
plt.plot(dm.index, dm['S2'], label='S2')
plt.plot(dm.index, dm['S3'], label='S3')
plt.plot(dm.index, dm['S4'], label='S4')

plt.scatter(ugleap.index, ugleap['S2'], label='S2')
plt.scatter(ugleap.index, ugleap['S3'], label='S3')
plt.scatter(ugleap.index, ugleap['S4'], label='S4')
plt.legend()
```

```python
plt.figure(figsize=(10,8))

dm = pd.read_csv("DM_MAPK_trajectory.csv", header=0)  
nrm = pd.read_csv("NRM_MAPK_trajectory.csv", header=0)
rssa = pd.read_csv("RSSA_MAPK_trajectory.csv", header=0)
ussa = pd.read_csv("USSA_MAPK_trajectory.csv", header=0)
unrm = pd.read_csv("UNRM_MAPK_trajectory.csv", header=0)
uleap = pd.read_csv("ULEAP_MAPK_trajectory.csv", header=0)

plt.plot(dm.index, dm['K'], color='red', label='DM')
plt.plot(dm.index, dm['Kp'], color='red')
plt.plot(dm.index, dm['Kpp'], color='red')
plt.scatter(nrm.index, nrm['K'], s=200, edgecolor='c', facecolors='none', label='NRM')
plt.scatter(nrm.index, nrm['Kp'], s=200, edgecolor='c', facecolors='none')
plt.scatter(nrm.index, nrm['Kpp'], s=200, edgecolor='c', facecolors='none')
plt.scatter(rssa.index, rssa['K'], s=150, edgecolor='m', facecolors='none', label='RSSA')
plt.scatter(rssa.index, rssa['Kp'], s=150, edgecolor='m', facecolors='none')
plt.scatter(rssa.index, rssa['Kpp'], s=150, edgecolor='m', facecolors='none')
plt.scatter(ussa.index, ussa['K'], s=100, edgecolor='y', facecolors='none', label='USSA')
plt.scatter(ussa.index, ussa['Kp'], s=100, edgecolor='y', facecolors='none')
plt.scatter(ussa.index, ussa['Kpp'], s=100, edgecolor='y', facecolors='none')
plt.scatter(unrm.index, unrm['K'], s=50, edgecolor='g', facecolors='none', label='RNRM')
plt.scatter(unrm.index, unrm['Kp'], s=50, edgecolor='g', facecolors='none')
plt.scatter(unrm.index, unrm['Kpp'], s=50, edgecolor='g', facecolors='none')
plt.scatter(uleap.index, uleap['K'], s=20, edgecolor='b', facecolors='none', label='RLEAP')
plt.scatter(uleap.index, uleap['Kp'], s=20, edgecolor='b', facecolors='none')
plt.scatter(uleap.index, uleap['Kpp'], s=20, edgecolor='b', facecolors='none')
        
plt.legend()
plt.xlabel("Time", fontdict=fontdict)
plt.ylabel("Population", fontdict=fontdict)
plt.savefig("MAPK_trajectory.pdf", dpi=600)
```

```python
bench = pd.read_table("MAPK_benchmark.txt", header=None)
dm = bench[0:100].values
nrm = bench[100:200].values
rssa = bench[200:300].values
ussa = bench[300:400].values
unrm = bench[400:500].values
plt.figure(figsize=(10,8))
plt.bar("DM", dm.sum()/1000)
plt.bar("NRM", nrm.sum()/1000)
plt.bar("RSSA", rssa.sum()/1000)
plt.bar("RNRM", unrm.sum()/1000)
plt.bar("USSA", ussa.sum()/1000)
plt.ylabel("CPU times (s)", fontdict=fontdict)
plt.savefig("MAPK_bench.pdf", dpi=600)
```
