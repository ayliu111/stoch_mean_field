```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
fontdict = {'family': 'Times New Roman', 'size': 20}
import matplotlib.transforms as mtransforms
from scipy.special import comb
```

A better approximation for the logarithm of a factorial can be found by using $\log n! \approx n \log n - n$. Interestingly, the additional terms in the approximation of the binomial coefficient cancel out, and the result is the same as if you used the simpler approximation $\log n! \approx n\log n$:

$$\begin{align} \log {n\choose m} & = \log \frac{n!}{m!(n-m)!} \\ & = \log n! - \log m! - \log (n-m)! \\ & \approx n \log n - n - m \log m + m - (n-m) \log (n-m) + n-m \\ & = n \log n - m \log m - (n - m) \log (n-m) \end{align}$$

An even better approximation uses more terms of Stirling's approximation, giving $\log n! \approx (n+\tfrac{1}{2})\log n - n + \tfrac{1}{2}\log 2\pi$ and hence

$$\begin{align} \log {n\choose m} &\approx (n+\tfrac{1}{2})\log n - (m+\tfrac{1}{2})\log m - (n-m+\tfrac{1}{2})\log (n-m) - \tfrac{1}{2}\log 2\pi \\ & = n\log n - m \log m - (n-m)\log(n-m) + \cdots \\ & \qquad +\tfrac{1}{2} (\log n - \log m - \log (n-m) - \log 2\pi) \end{align}$$

where the last term is the correction from using more terms of Stirling's approximation.

```python
n = np.arange(10, 200)
k = 3 * np.ones_like(n)
c = comb(n, k)
plt.plot(n, c, 'r', label="accurate")
plt.plot(n, np.power(n/k, k), 'b', label="lb")
plt.plot(n, np.power(np.e * n / k, k), 'y', label="ub")
plt.plot(n, np.exp(n * np.log(n) - k * np.log(k) - (n - k) * np.log(n - k)), 'g', label="stirling")
plt.legend()
```

```python
k = np.arange(2, 10)
n = 100 * np.ones_like(k)
c = comb(n, k)
plt.plot(k, np.log(c), label="accurate")
plt.plot(k, np.log(np.power(n / k, k)), label="lb")
plt.plot(k, np.log(np.power(np.e * n / k, k)), label="ub")
plt.plot(k, k * np.power(c, 1/k) / n, label="ub")
plt.legend()
```

```python
plt.figure(figsize=(10,8))
k = np.arange(1, 5)
s = np.array([1, 2, 3, 4, 2, 4, 6, 8])
ns = np.arange(100, 5000, 100)
for i in k:
    x1 = []
    x2 = []
    for j in ns:
        n = j * np.ones_like(s)
        t1 = comb(n+s, i)/comb(n, i)
        z = comb(j+s.mean(), i)/comb(j,i)
        x1.append(np.prod(t1))
        x2.append(np.power(z, t1.size))
    plt.plot(ns, np.log(x1), 'o', ms=5, label=rf"$k$={i}")
    plt.plot(ns, np.log(x2), 'r')
plt.xlabel(r"$n$", fontsize=20)
plt.ylabel("Log Weight", fontsize=20)
plt.legend()
plt.title("Approximate propensities ratio with middel population")
```

```python
plt.figure(figsize=(10,8))
k = np.arange(1, 5)
s = np.array([1, 2, 3, 4, 2, 4, 6, 8])
ns = np.arange(100, 5000, 100)
for i in k:
    x1 = []
    x2 = []
    for j in ns:
        n = j * np.ones_like(s)
        t1 = comb(n+s, i) - comb(n, i)
        z = comb(j+s.mean(), i) - comb(j,i)
        x1.append(np.sum(t1))
        x2.append(z * t1.size)
    plt.plot(ns, np.log(x1), 'o', ms=5, label=rf"$k$={i}")
    plt.plot(ns, np.log(x2), 'r')
plt.xlabel(r"$n$", fontsize=20)
plt.ylabel("Log Weight", fontsize=20)
plt.legend()
plt.title("Approximate propensities difference with middel population")
```

Thus we can approximate the weight by it middle population propensity. 
The propensity function for reaction $j$ is

$$h_j(\mathbf{X}) = r_j \prod_k {n_{jk} \choose \nu_{jk}}$$

where $r_j$ is the rate, $n_{jk}$ is the number of molecules for $k$th reactant, $\nu_{jk}$ is the stoichiometric coefficient for $k$th reactant, and the weight for a leap $l$ with $s$ steps is 

$$w_{jl} = \prod_s \prod_k \frac{{n_{jk} + n_{jks} \choose \nu_{jk}}}{{n_{jk} \choose \nu_{jk}}}$$

where $n_{jks}$ is the difference of number of molecules for $k$th reactant of reaction $j$ of the $s$ step of the leap. 
We have shown in the figure that we can approximate the ratio of combinatorial number as

$$\prod_s \frac{{n_0 + n_s \choose k}}{{n_0 \choose k}} \approx \left[\frac{{n_0 + \left<n_s\right>_s \choose k}}{{n_0 \choose k}}\right]^s$$

Thus we can approximate the weight of propensity by 

$$w_{jl} =\prod_k \left[\frac{{n_{jk} + \left<n_{jks}\right>_s \choose \nu_{jk}}}{{n_{jk} \choose \nu_{jk}}} \right]^s = \left[\prod_k \frac{{n_{jk} + \left<n_{jks}\right>_s \choose \nu_{jk}}}{{n_{jk} \choose \nu_{jk}}} \right]^s$$

which is just

$$w_{jl} = \left[\frac{h_j(\left<\mathbf{X}\right>_l)}{h_j(\mathbf{X}_0)}\right]^s$$

The weight for time difference is 

$$w_{tl} = \exp \left\{- \sum_s \left[ \sum_j r_j \left(\prod_k {n_{jk} + n_{jks} \choose \nu_{jk}} - \prod_k {n_{jk}\choose \nu_{jk}}\right)\right] \Delta t_s \right\}$$

We can approximate the difference of combinatorial number as 

$$\sum_s \left(\prod_k {n_{jk} + n_{jks} \choose \nu_{jk}} - \prod_k {n_{jk}\choose \nu_{jk}}\right) \Delta t_s \approx s \cdot \left(\prod_k {n_{jk} + \left<n_{jks}\right>_s \choose \nu_{jk}} - \prod_k {n_{jk}\choose \nu_{jk}}\right) \left<\Delta t_s\right>_s$$

where $\left<\Delta t_s\right>_s$ is the mean time interval. Thus weight is approximated by 

$$w_{tl} = \exp \left\{- s \cdot \left[ \sum_j r_j \left(\prod_k {n_{jk} + \left<n_{jks}\right>_s \choose \nu_{jk}} - \prod_k {n_{jk}\choose \nu_{jk}}\right)\right] \left<\Delta t_s\right>_s \right\}$$

which is just 

$$w_{tl} = \exp \left\{- s \cdot \left(a(\left<\mathbf{X}\right>_l) - a(\mathbf{X}_0)\right) \left<\Delta t_s\right>_s \right\}$$

where $$a(\mathbf{X}) = \sum_j h_j(\mathbf{X})$$. If we chose $\left<\Delta t_s\right>_s = 1 / a(\left<\mathbf{X}\right>_l)$, we have finally

$$w_{tl} = \exp \left\{- s \cdot \left(1 - \frac{a(\mathbf{X}_0)}{a(\left<\mathbf{X}\right>_l)} \right)\right\}$$


Therefore the total weight is approximated by

$$w_l = \prod_j \left[\frac{h_j(\left<\mathbf{X}\right>_l)}{h_j(\mathbf{X})}\right]^{s_{lj}} \exp \left\{s \cdot \left(\frac{a(\mathbf{X}_0)}{a(\left<\mathbf{X}\right>_l)} - 1 \right)\right\}$$

Thus the weight of the leap can be approximated by computing the propensities of the leap-mean state $\left<\mathbf{X}\right>_l$ and the starting state $\mathbf{X}_0$. Computational cost for calculating step-wise weight is thus saved.
