```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
fontdict = {'family': 'Times New Roman', 'size': 20}
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']
import matplotlib.transforms as mtransforms
```

$$
  \overline{\gamma}_j = \sup_n \overline{a}_j/a_j = \prod_{i=1}^{N_j} \frac{\overline{n}(\overline{n}-1) \ldots (\overline{n} - \nu_{ij} + 1)}{n(n-1) \ldots (n - \nu_{ij} + 1)} 
$$
$$
\underline{\gamma}_j = \inf_n \underline{a}_j/a_j = \prod_{i=1}^{N_j} \frac{\underline{n}(\underline{n}-1) \ldots (\underline{n} - \nu_{ij} + 1)}{n(n-1) \ldots (n - \nu_{ij} + 1)} 
$$
where $\overline{n} = \left\lfloor (1+\bound_factors)n \right\rfloor$, $\underline{n} = \left\lceil (1-\bound_factors)n \right\rceil$, and $n = \left\lceil 1/\bound_factors \right\rceil$.

$$
\inf_n p_{j, \text{early}} = \underline{\gamma_j}/\overline{\gamma_j} = \prod_{i=1}^{N_j} \frac{\underline{n}(\underline{n}-1) \ldots (\underline{n} - \nu_{ij} + 1)}{\overline{n}(\overline{n}-1) \ldots (\overline{n} - \nu_{ij} + 1)}
$$
$$
\inf_n p_{j, \text{early}} \ge \prod_{i=1}^{N_j} \left[\frac{\underline{n}_j - \nu_{ij} + 1}{\overline{n}_j - \nu_{ij} + 1}\right]^{\nu_{ij}}  \ge \prod_{i=1}^{N_j} \left[\frac{\underline{n}_j - \nu_{j} + 1}{\overline{n}_j - \nu_{j} + 1}\right]^{\nu_{ij}} = \left[\frac{\underline{n}_j - \nu_{j} + 1}{\overline{n}_j - \nu_{j} + 1}\right]^{\sum_i^{N_j} \nu_{ij} = S} \ge p_b
$$
where $\nu_{j} = \arg \max_i\{\nu_{ij}\}$, and $\overline{n}_j = \left\lfloor (1+\delta_j)n \right\rfloor$, $\underline{n}_j = \left\lceil (1-\delta_j)n \right\rceil$, since each channel will be assign a different $\delta_j$.
$$
		\frac{\left\lceil (1-\delta_j)\left\lceil 1/\delta_j \right\rceil \right\rceil - \nu_{j} + 1}{\left\lfloor (1+\delta_j)\left\lceil 1/\delta_j \right\rceil \right\rfloor - \nu_{j} + 1} \ge \frac{(1-\delta_j)\left\lceil 1/\delta_j \right\rceil - \nu_{j}}{(1+\delta_j)\left\lceil 1/\delta_j \right\rceil - \nu_{j}} \ge \frac{(1-\delta_j) (1/\delta_j) - \nu_{j} + 1}{(1+\delta_j) (1/\delta_j) - \nu_{j} + 1} = p_b^{1/S}
$$
$$
		\delta_j = \frac{1-p_b^{1/S}}{(1-p_b^{1/S})\nu_j + 2p_b^{1/S}}	
$$

```python
def combinatorial(n, k):
    """calculate combinatorial number, n chose k"""
    out = 1
    for i in range(1, k+1):
        out *= (n-(i-1))/i
    return out
```

```python
def bound_factor(n1, n2, k):
    """calculate bound factor, for N_j = 1, n1 and n2 are the input number
        e.g. n1 = np.ceil((1-bound_factors)*n), n2 = n, k is the stoichiometric number"""
    out = 1
    for i in range(1, k+1):
        out *= (n1-(i-1))
        out /= (n2-(i-1))
    return out
```

```python
plt.figure(figsize=(10,10))
bound_factors = np.linspace(0.01, 0.9, 100)
n = np.ceil(1/bound_factors)
v = 2
plt.plot(bound_factors, (np.ceil((1-bound_factors)*n)-v+1)/(np.floor((1+bound_factors)*n)-v+1), label='1')
plt.plot(bound_factors, ((1-bound_factors)*n-v+1)/((1+bound_factors)*n-v+1), label='2')
plt.plot(bound_factors, ((1-bound_factors)*1/bound_factors-v+1)/((1+bound_factors)*1/bound_factors-v+1), label='3')
plt.legend()
plt.title(r"Calculating $\bound_factors$ from bound probability, showing the two inequalities", fontdict=fontdict)
plt.xlabel(r"$\bound_factors$", fontdict=fontdict)
plt.ylabel("Accept probability", fontdict=fontdict)
```

```python
plt.figure(figsize=(12, 10))
prob_lb = 0.9
N = 10000
for s in range(1, 8):
    pbs = prob_lb**(1/s)
    bound_factors = (1-pbs)/((1-pbs)*s+2*pbs)
    ueta = 1+bound_factors
    leta = 1-bound_factors
    n = np.arange(s+1, N)
    ly = combinatorial(np.ceil(leta*n), s)/combinatorial(n, s)
    uy = combinatorial(np.floor(ueta*n), s)/combinatorial(n, s)
    plt.plot(n, ly/uy, label=fr's={s}, $\bound_factors$={bound_factors:.3f}', color=colors[(s-1)%len(colors)])
plt.legend()
plt.hlines(y=prob_lb, xmin=-300, xmax=N+300, linestyles='dashed', color='r')
plt.xlabel("Population", fontdict=fontdict)
plt.ylabel("Acceptance probability", fontdict=fontdict)
plt.title(r"Calculate $\bound_factors$ from bound acceptance probability", fontdict=fontdict)
```

```python

plt.figure(figsize=(10,10))
bound_factors = 0.03
ueta = 1+bound_factors
leta = 1-bound_factors
N = 100
for s in range(1, 6):
    n = np.arange(s+1, N)
    ly = combinatorial(np.ceil(leta*n), s)/combinatorial(n, s)
    uy = combinatorial(np.floor(ueta*n), s)/combinatorial(n, s)
    plt.plot(n, uy, label=f's={s}', color=colors[s-1])
    plt.plot(n, ly, label=f's={s}', color=colors[s-1])

plt.vlines(x=np.ceil(1/bound_factors), ymin=0.8, ymax=1.2)
plt.title(r"The critical point is always at $\left\lceil 1/\bound_factors \right\rceil$", fontdict=fontdict)
plt.xlabel(r"Population", fontdict=fontdict)
plt.ylabel(r"Propensity factor", fontdict=fontdict)
```

```python
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']
fig, axes = plt.subplots(1, 3, figsize=(30, 8))
bound_factors = 0.01
ueta = 1+bound_factors
leta = 1-bound_factors
N = 10000
for s in range(1, 6):
    n = np.arange(s+1, N)
    ly = combinatorial(np.ceil(leta*n), s)/combinatorial(n, s)
    uy = combinatorial(np.floor(ueta*n), s)/combinatorial(n, s)
    plt.subplot(131)
    plt.plot(n, uy, label=f's={s}', color=colors[s-1])
    plt.plot(n, ly, label=f's={s}', color=colors[s-1])
    plt.subplot(132)
    plt.plot(n, ly/uy, label=f's={s}', color=colors[s-1])
    n_min = np.ceil(1/bound_factors)
    lby = bound_factor(np.ceil(leta*n_min), n_min, s)*np.ones(n.size)
    uby = bound_factor(np.floor(ueta*n_min), n_min, s)*np.ones(n.size)
    plt.subplot(131)
    plt.plot(n, lby, color='black', linestyle='dashed')
    plt.plot(n, uby, color='black', linestyle='dashed')
    plt.subplot(132)
    plt.plot(n, lby/uby, color='black', linestyle='dashed')
plt.legend()
plt.subplot(131)
plt.xlabel("Population", fontdict=fontdict)
plt.ylabel("Propensity ratio", fontdict=fontdict)
trans = mtransforms.ScaledTranslation(-50/72, 7/72, fig.dpi_scale_trans)
plt.text(0.0, 1.0, 'A', transform=axes[0].transAxes + trans,
            fontsize=20, verticalalignment='bottom', fontfamily='serif',
            bbox=dict(facecolor='1.0', edgecolor='none', pad=3.0))
plt.subplot(132)
plt.xlabel("Population", fontdict=fontdict)
plt.ylabel("Early acceptance probability", fontdict=fontdict)
trans = mtransforms.ScaledTranslation(-50/72, 7/72, fig.dpi_scale_trans)
plt.text(0.0, 1.0, 'B', transform=axes[1].transAxes + trans,
            fontsize=20, verticalalignment='bottom', fontfamily='serif',
            bbox=dict(facecolor='1.0', edgecolor='none', pad=3.0))
# plt.axvline(np.ceil(1/bound_factors), color='black', linestyle='dashed')
plt.subplot(133)
prob_lb = 0.9
N = 10000
for s in range(1, 6):
    pbs = prob_lb**(1/s)
    bound_factors = (1-pbs)/((1-pbs)*s+2*pbs)
    ueta = 1+bound_factors
    leta = 1-bound_factors
    n = np.arange(s+1, N)
    ly = combinatorial(np.ceil(leta*n), s)/combinatorial(n, s)
    uy = combinatorial(np.floor(ueta*n), s)/combinatorial(n, s)
    plt.plot(n, ly/uy, label=fr's={s}, $\bound_factors$={bound_factors:.3f}', color=colors[(s-1)%len(colors)])
plt.legend()
plt.hlines(y=prob_lb, xmin=-300, xmax=N+300, linestyles='dashed', color='k')
plt.xlabel("Population", fontdict=fontdict)
plt.ylabel("Acceptance probability", fontdict=fontdict)
trans = mtransforms.ScaledTranslation(-50/72, 7/72, fig.dpi_scale_trans)
plt.text(0.0, 1.0, 'C', transform=axes[2].transAxes + trans,
            fontsize=20, verticalalignment='bottom', fontfamily='serif',
            bbox=dict(facecolor='1.0', edgecolor='none', pad=3.0))

plt.tight_layout()

plt.savefig("early_accept_prob_s.pdf", dpi=600, bbox_inches=0, pad_inches=0)
```
