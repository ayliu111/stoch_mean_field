```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
fontdict = {'family': 'Times New Roman', 'size': 20}
import matplotlib.transforms as mtransforms
```

```python
def plot_hist(data, mtype, name):
    data = data.astype(int)
    ub = data.max()+10
    lb = data.min()-10
    count = np.zeros(ub-lb+1)
    for i in data:
        count[i-lb] += 1
    plt.plot(np.arange(lb, ub+1), count/count.sum(), mtype, label=name)
```

### The birth-death model

A $\to$ $\emptyset$, $k_1 = 0.1$

$\emptyset$ $\to$ A, $k_2 = 1$

Initial A(0) = 0

For the chemical master equation, the flow in terms are 

$k_1(n+1)P_{n+1}+k_2P_{n-1}$

the flow out terms are

$-k_1 n P_n - k_2 P_n$

Thus the chemical master equation is 
$$
\frac{dP_n}{dt} = k_1(n+1)P_{n+1}+k_2P_{n-1} -k_1 n P_n - k_2 P_n
$$
Define the asymptotic stationary distribution 
$$
\phi(n) = \lim_{t \to \infty} p_n(t)
$$
We then have 
$$
\frac{d\phi(n)}{dt} = 0
$$
So we have the induction relations
$$
0 = k_1 \phi(1) - k_2 \phi(0)
$$
for $n = 0$
$$
0 = k_1(n+1)\phi(n+1) + k_2 \phi(n-1) - k_1 n \phi(n) - k_2 \phi(n)
$$
for $n \ge 1$
They are
$$
\phi(1) = \frac{k_2}{k_1} \phi(0)
$$
for $n = 0$, and 
$$
\phi(n+1) = \frac{1}{k_1(n+1)}[k_1n\phi(n) + k_2 \phi(n) - k_2 \phi(n-1)]
$$
Assign $\phi(n) = 1$, we can get the relative probability weight, and divide all weights with the sum of weights is equivalent to normalization. 

```python
plt.figure(figsize=(10,8))

dm = pd.read_csv("DM_birth_death_trajectory_100.csv", header=0)  
nrm = pd.read_csv("NRM_birth_death_trajectory_100.csv", header=0)
rssa = pd.read_csv("RSSA_birth_death_trajectory_100.csv", header=0)
ussa = pd.read_csv("USSA_birth_death_trajectory_100.csv", header=0)
unrm = pd.read_csv("UNRM_birth_death_trajectory_100.csv", header=0)
uleap = pd.read_csv("ULEAP_birth_death_trajectory_100.csv", header=0)

plt.plot(dm.index, dm['A'], color='r', label='DM')
plt.scatter(nrm.index, nrm['A'], s=200, edgecolor='c', facecolors='none', label='NRM')
plt.scatter(rssa.index, rssa['A'], s=150, edgecolor='m', facecolors='none', label='RSSA')
plt.scatter(ussa.index, ussa['A'], s=100, edgecolor='y', facecolors='none', label='USSA')
plt.scatter(unrm.index, unrm['A'], s=50, edgecolor='g', facecolors='none', label='UNRM')
plt.scatter(uleap.index, uleap['A'], s=20, edgecolor='b', facecolors='none', label='ULEAP')
        
plt.legend()
plt.tight_layout()
plt.xlabel("Time", fontdict=fontdict)
plt.ylabel("A Population", fontdict=fontdict)
plt.savefig("birth_death_trajectory.pdf", dpi=600)
```

### The isomerization model

A $\to$ B, $k_1 = 10$

B $\to$ A, $k_2 = 10$

Initial A(0) = 100, B(0) = 100

The CME is still a one parameter equation

CME:
$$
\frac{dP_n}{dt} = k_1(n+1)P_{n+1}+k_2(N-n+1)P_{n-1}-k_1nP_n-k_2(N-n)P_n
$$
for $n=0$, 
$$
\phi(1) = \frac{k_2N}{k_1} \phi(0)
$$
for $n \ge 0$, 
$$
\phi(n+1) = \frac{1}{k_1(n+1)} [k_1n\phi(n) + k_2(N-n)\phi(n) - k_2(N-n+1)\phi(n-1)]
$$

### Sequential model

A $\to$ B, $k_1 = 1$

B $\to$ C, $k_2 = 1$

$(m+1, n-1, l) \to (m, n, l)$

$(m, n+1, l-1) \to (m, n, l)$

$(m, n, l) \to (m-1, n+1, l)$

$(m, n, l) \to (m, n-1, l+1)$

So the master equation is 
$$
\frac{dP(m, n, l)}{dt} = k_1 (m+1) P(m+1, n-1, l) + k_2 (n+1) P(m, n+1, l-1) - (k_1 m + k_2 n) P(m, n, l)
$$
The calculation can be done numerically

Since the total molecular number is set, it can be reduced into a 2 variable differential equation
$$
\frac{dP(m, n)}{dt} = k_1 (m+1) P(m+1, n-1) + k_2 (n+1) P(m, n+1) - (k_1 m + k_2 n) P(m, n)
$$

### Schlogl model

B1 + 2X $\to$ 3X, $c_1 = 3e-7$

3X $\to$ B1 + 2X, $c_2 = 1e-4$

B2 $\to$ X, $c_3 = 1e-3$

X $\to$ B2, $c_4 = 3.5$

B1 = $N_1$, and B2 = $N_2$ are constant

The master equation is 

$$
\frac{dP_n}{dt} = c_1 N_1 \frac{1}{2} (n-1) (n-2) P_{n-1} + \frac{1}{6} (n+1) n (n-1) c_2 P_{n+1} + c_3 N_2 P_{n-1} + c_4 (n+1) P_{n+1} - c_1N_1 \frac{1}{2}n (n-1) P_n - \frac{1}{6} n (n-1) (n-2) c_2 P_n - c_3 N_2 P_n - c_4 n P_n \\
= a P_{n-1} + b P_{n+1} - c P_n
$$

where 

$$
a = c_1 N_1 \frac{1}{2} (n-1) (n-2) + c_3 N_2 \\
b = \frac{1}{6} (n+1) n (n-1) c_2 +  c_4 (n+1) \\
c = c_1N_1 \frac{1}{2}n (n-1) + \frac{1}{6} n (n-1) (n-2) c_2 + c_3 N_2 + c_4 n
$$

$$
    \phi(1) = \frac{c}{b} \phi(0)
$$
for $n = 0$, and 
$$
\phi(n+1) = \frac{1}{b}[c\phi(n) -a \phi(n-1)]
$$

```python
plt.figure(figsize=(10,8))

p = []
c1 = 3e-7
c2 = 1e-4
c3 = 1e-3
c4 = 3.5
n1 = 1e5
n2 = 197000
N = 800
p.append(1)
n = 0
a = c1*n1/2*(n-1)*(n-2)+c3*n2
b = 1/6*(n+1)*n*(n-1)*c2 + c4*(n+1)
c = c1*n1/2*n*(n-1)+1/6*n*(n-1)*(n-2)*c2+c3*n2+c4*n

x = c/b*p[0]

p.append(x)
for n in range(1, N):
    a = c1*n1/2*(n-1)*(n-2)+c3*n2
    b = 1/6*(n+1)*n*(n-1)*c2 + c4*(n+1)
    c = c1*n1/2*n*(n-1)+1/6*n*(n-1)*(n-2)*c2+c3*n2+c4*n
    x = 1/b*(c*p[n]-a*p[n-1])
    p.append(x)
    
p = np.array(p)    
ix = np.arange(0, N)

plt.plot(ix, p[ix]/p.sum(), color='black', lw=2, label='CME')


dmh = pd.read_csv("DM_schlogl_histogram.csv", header=0)
nrmh = pd.read_csv("NRM_schlogl_histogram.csv", header=0)
rssah = pd.read_csv("RSSA_schlogl_histogram.csv", header=0)
ussah = pd.read_csv("USSA_schlogl_histogram.csv", header=0)
unrmh = pd.read_csv("UNRM_schlogl_histogram.csv", header=0)
uleaph = pd.read_csv("ULEAP_schlogl_histogram.csv", header=0)

dm = dmh["S1"].to_numpy()
nrm = nrmh["S1"].to_numpy()
rssa = rssah["S1"].to_numpy()
ussa = ussah["S1"].to_numpy()
unrm = unrmh["S1"].to_numpy()
uleap = uleaph["S1"].to_numpy()

plot_hist(dm, 'ro', 'DM')
plot_hist(nrm, 'c^', 'NRM')
plot_hist(rssa, 'm>', 'RSSA')
plot_hist(ussa, 'y<', 'USSA')
plot_hist(unrm, 'gv', 'UNRM')
plot_hist(uleap, 'b+', 'ULEAP')
plt.plot(ix, p[ix]/p.sum(), color='black', lw=2)

plt.xlabel("S1 population", fontdict=fontdict)
plt.ylabel("Probability", fontdict=fontdict)
plt.legend()
plt.savefig("schlogl_histogram.pdf", dpi=600)
```
